```python
import pandas as pd
df = pd.read_csv('10prog_4laptops.csv')
df.head
```

```python
pd.get_dummies(df)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc_data = enc.fit_transform(df[['Category']])
enc_data
```

```python
enc_data.toarray()
enc_df = pd.DataFrame(enc_data.toarray())
enc_df
```

```python
enc_df = pd.DataFrame(enc_data.toarray(), columns = ['2 in 1 Convertible',
'Gaming', 'Netbook', 'Notebook', 'Ultrabook', 'Workstation'])
enc_df
```

```python
df1 = df.join(enc_df) 
df1
```
