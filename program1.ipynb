```python
try:
    result=1/0
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter the denominator grater than 0")    
```

```python
#try:risky errors are typed here
#except:error handling
#else:when there are no error in the code else will be executed
#finally:conclusion type no matter if errors are there or not finally will be executed
#exeption means general error, it will handel all the error if it is not mentioned
#while printing it will take only first error, and ignores the second error
```

```python
try:
    result=1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter the denominator grater than 0")    
except Exception as ex1:
    print(ex1)
    print("Main exeption caught here")    
```
